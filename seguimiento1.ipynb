Ejercicio numero 4

```python
def calcularmamajuan  (edadjuan, edadalberto, edadana):
    edadmama = edadjuan + edadalberto + edadana
    return  edadmama
def calcularalberto (edadjuan):
    edadalberto = (edadjuan/3)*2
    return edadalberto
def calcularana (edadjuan):
    edadana = (edadjuan/3)*4
    return edadana

edadjuan = int(input("Cual es la edad de juan?"))
edadalberto = calcularalberto(edadjuan)
edadana = calcularana(edadjuan)
edadmama = calcularmamajuan(edadjuan, edadalberto, edadana)

print(f"Alberto tiene: {edadalberto}")
print(f"Ana tiene: {edadana}")
print(f"La mamá tiene: {edadmama}")
```

```python
#Ejercicio resuelto numero 4 orientado a objetos
class familia:#innit es quien hace que se ejecute la funcion cuando se nombra el objeto
    def __init__(self, edadjuan): #tiene en cuenta el self, el espacio de memoria o cookie y el parametro
        #lo que el codigo debe saber, la funcion es init, los objetos los self.
        self.juan = edadjuan
        self.alberto = (edadjuan/3)*2
        self.ana = (edadjuan/3)*4

    def calcularmama(self):#cuando utilizo la siguiente funcion sin parametros externos es metodo
        return self.juan + self.alberto + self.ana

#programa principal
inputedad = int(input("¿Cuál es la edad de Juan?:"))
#creamos el objeto a partir de la clase
mifamilia = familia(inputedad)#le damos el parametro

print(f"alberto tiene:{mifamilia.alberto:.1f}")#los dos puntos despues de alberto es para decir que hay una regla de formato
#recordad las comillas del fstring y el 1f = un flotante
print(f"ana tiene:{mifamilia.ana:.1f}")
print(f"la edad de mamá es:{mifamilia.calcularmama():.1f}")
```

Ejercicio Propuesto No 12

```python
'''Un empleado trabaja 48 horas en la semana a razón de $5.000 hora. El porcentaje de
retención en la fuente es del 12,5% del salario bruto. Se desea saber cuál es el salario bruto,
la retención en la fuente y el salario neto del trabajador. '''
class salario:
    def __init__(self,horas,precio):
        self.horas=horas
        self.precio=precio

    def salario_bruto(self):
        return self.horas*self.precio #al llamar a self.horas o precio no utilizo () porque es un atributo, algo estatico
    def retencion(self):
        return self.salario_bruto()*0.125 #al llamar a self.salario_bruto utilizo () porque es un metodo, parte de la class
    def salario_neto(self):
        return self.salario_bruto()-self.retencion()

horas = int(input("Ingrese las horas trabajadas: "))
precio = int(input("Ingrese el precio por hora: "))

salario = salario(horas,precio)#recordar darle una variable a la class y sus parametros
print("El salario bruto es: ",salario.salario_bruto())
print("La retencion es: ",salario.retencion())
print("El salario neto es: ",salario.salario_neto())
```

Ejercicio propuesto No 14

```python
'''Elabore un algoritmo que lea un número y obtenga su cuadrado y su cubo.'''
class calcularcu:
    def __init__(self, num):#se me habia olvidado el self del init
        self.num = num

    def cuadrado(self):
        return self.num**2
    def cubo(self):
        return self.num**3
numero = float(input("ingrese el numero"))
calcular = calcularcu(numero)
print("Cuadrado: ", calcular.cuadrado())
print("cubo: ", calcular.cubo())
```

```python
'''Me di cuenta que cuando las funciones no depende de otras es mejor utilizar staticmethod para no almacenar nada en la memoria''' 
class calcularcu:
    @staticmethod#con staticmethod no es necesario el self
    def cuadrado(num):
        return num**2
    @staticmethod
    def cubo(num):
        return num**3
numero = float(input("ingrese el numero"))
#como no es necesario el self se llama a la class directamente
print("Cuadrado: ", calcularcu.cuadrado(numero))
print("cubo: ", calcularcu.cubo(numero))
```

Ejercicio propuesto No 17

```python
'''dado el radio de un círculo. Haga un algoritmo que obtenga el área del círculo y la longitud
de la circunferencia.'''
from math import pi #primero el from y luego el import

class calcularcirculo:
    @staticmethod
    def area(radio):
        return pi*radio**2
    @staticmethod
    def perimetro(radio):
        return 2*pi*radio

radio = float(input("Ingrese el radio del círculo: "))

print(f"El área del círculo es: {calcularcirculo.area(radio):.3f}")#el :.3f solo funciona en fstrings
print(f"La longitud de la circunferencia es:{calcularcirculo.perimetro(radio):.3f}")
```
